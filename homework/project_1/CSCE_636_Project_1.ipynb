```python
import pickle
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
#mount the drive and define the file locations
#anyone can access this link if needed to verify results
#the project will be saved on my google drive at https://drive.google.com/drive/folders/1wIap3EZs3pt82AGXG1hpQIFp7hLMsJfT?usp=drive_link
from google.colab import drive
drive.mount('/content/drive')

TRAIN_DATA_PATH   = '/content/drive/MyDrive/CSCE_636/project_1/CSCE-636-Project-1-Train-n_k_m_P'
TRAIN_HEIGHT_PATH = '/content/drive/MyDrive/CSCE_636/project_1/CSCE-636-Project-1-Train-mHeights'
MODEL_SAVE_PATH   = '/content/drive/MyDrive/CSCE_636/project_1/best_model.pth'
NUM_EPOCHS        = 200
BATCH_SIZE        = 256
LEARNING_RATE     = 1e-3
```

```python
#As a sanity check make sure a GPU is selected that can handle this computation
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')
```

```python
#Load the data
with open(TRAIN_DATA_PATH, 'rb') as f:
    train_data = pickle.load(f)
with open(TRAIN_HEIGHT_PATH, 'rb') as f:
    train_heights = pickle.load(f)
#print the length of our data to determine that the upload happend correctly
print(f'Total samples: {len(train_data)}')
```

```python
def encode_sample(sample):
  #P is a 2D array subset of G, has k rows and varies between 4,5,6 and n-k columns
  #n is the total length of the codeword which is C and the paper says it is always 9
  # c= xG
  # m controls which ranked position you compare against when computing the height, varies between 2 and n-k
  #Rank in this case means in order from highest absolute value to lowest
  #divide the rank by position 0 this gives you h_m(c)
  # max of h_m(c) is the height for that specific code word
  n,k,m, P = sample
  p_flat = P.flatten()
  #this is from a formula in the given paper k*(n-k)
  p_padded = np.zeros(20)
  #put the actual p values up front and then drop in 0s
  p_padded[:len(p_flat)] = p_flat
  p_padded = p_padded / 100.0  # normalize to [-1, 1]
  #dividing by 9 to match the normalized p values
  features = np.concatenate([[n / 9.0, k / 9.0, m / 9.0], p_padded])
  return features.astype(np.float32)

X = np.array([encode_sample(s) for s in train_data])
# m-height values from the training data converted to log2space
y = np.array([np.log2(h) for h in train_heights], dtype=np.float32)

print(f'Feature shape: {X.shape}')
print(f'Target (log2 height) range: {y.min():.2f} to {y.max():.2f}')

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=42)
print(f'Train: {len(X_train)}, Val: {len(X_val)}')
```

```python
class MHeightDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]
#train set
train_loader = DataLoader(MHeightDataset(X_train, y_train), batch_size=BATCH_SIZE, shuffle=True)
#Validation set
val_loader   = DataLoader(MHeightDataset(X_val,   y_val),   batch_size=BATCH_SIZE, shuffle=False)
```

```python
#start with a 256,512,256,128,1 and see what happens
class MHeightDNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(23, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(0.2),

            nn.Linear(256, 512),
            nn.ReLU(),
            nn.BatchNorm1d(512),
            nn.Dropout(0.2),

            nn.Linear(512, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(0.1),

            nn.Linear(256, 128),
            nn.ReLU(),

            nn.Linear(128, 1)
        )

    def forward(self, x):
        return self.net(x).squeeze(-1)
#create an instance of the model and move it to the GPU if available
model = MHeightDNN().to(device)
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'Total trainable parameters: {total_params:,}')
```

```python
#helper function to calculate the loss
# delta = (log2(y) - log2(y_hat))^2 = (y_true - y_pred)^2
def custom_loss(pred, true):
    return torch.mean((true - pred) ** 2)

optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)
```

```python
best_val_loss = float('inf')
train_losses, val_losses = [], []

for epoch in range(NUM_EPOCHS):
    model.train()
    train_loss = 0.0
    for X_batch, y_batch in train_loader:
        #makes sure the calculation happens on the GPU
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        loss = custom_loss(model(X_batch), y_batch)
        #how much did each weight in the network contribute to the loss
        loss.backward()
        #update the weights
        optimizer.step()
        train_loss += loss.item()
    #average loss across all batches in the epoch
    train_loss /= len(train_loader)

    model.eval()
    val_loss = 0.0
    #no_grad does not track the gradients just evaluates the model
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            val_loss += custom_loss(model(X_batch), y_batch).item()
    val_loss /= len(val_loader)

    scheduler.step(val_loss)
    train_losses.append(train_loss)
    val_losses.append(val_loss)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), MODEL_SAVE_PATH)

    if epoch % 10 == 0 or epoch == NUM_EPOCHS - 1:
        print(f'Epoch {epoch:3d} | Train: {train_loss:.4f} | Val: {val_loss:.4f} | Best: {best_val_loss:.4f}')

print(f'\nTraining complete. Best val loss: {best_val_loss:.4f}')
```

```python
#plot the loss to evaluate the model if it is over or underfitting
plt.figure(figsize=(10, 4))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses,   label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss (log2 space)')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('training_curves.png')
plt.show()
```

```python
#test model performance
def predict(test_data, model_path=MODEL_SAVE_PATH):
    """
    Predict m-heights for a list of (n, k, m, P) samples.

    Args:
        test_data: list of [n, k, m, P] samples
        model_path: path to saved model weights

    Returns:
        numpy array of predicted m-heights (>= 1.0)
    """
    model = MHeightDNN().to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()

    X_test = torch.tensor(
        np.array([encode_sample(s) for s in test_data]),
        dtype=torch.float32
    )

    with torch.no_grad():
        log2_preds = model(X_test.to(device)).cpu().numpy()

    # Convert from log2 space, clamp to >= 1 as required
    return np.maximum(2.0 ** log2_preds, 1.0)
```

```python
#Test the model performance with a validation set
def validation_prediction(train_data, train_heights):
    _, val_idx = train_test_split(np.arange(len(train_data)), test_size=0.1, random_state=42)
    val_samples = [train_data[i] for i in val_idx]
    val_true_heights = [train_heights[i] for i in val_idx]

    val_predictions = predict(val_samples)

    for i in range(5):
        print(f'True: {val_true_heights[i]:.4f} | Predicted: {val_predictions[i]:.4f}')

    print(f'\nOverall average cost: {np.mean((np.log2(val_true_heights) - np.log2(val_predictions))**2):.4f}')


validation_prediction(train_data, train_heights)
```
