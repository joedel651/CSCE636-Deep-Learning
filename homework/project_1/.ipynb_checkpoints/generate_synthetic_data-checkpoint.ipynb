# Synthetic Data Generation for m-Height DNN

## Why This Notebook Exists

This notebook diagnoses **why the `augmented_training_data_300k.pkl` file hurt model performance** and then generates corrected synthetic training data that matches the format expected by the existing model pipeline.

---

## Root Cause Analysis: Why the Augmented Data Made Things Worse

After inspection, the augmented data had **four distinct problems**:

### Problem 1: Format Mismatch (Critical)
The notebook's `encode_sample` function does:
```python
n, k, m, P = sample
```
This **requires** `sample` to be a list of exactly 4 elements: `[n, k, m, P_matrix]`.

The augmented `.pkl` file stores dictionaries: `{'n':..., 'k':..., 'm':..., 'P':..., 'h':...}`. When you try to unpack a dict with 5 keys into 4 variables, Python throws `ValueError: too many values to unpack`. If someone worked around this by extracting only the list side, the `h` labels (heights) were embedded inside the dict — not in a separate list — so the model would have been trained with **misaligned X/y pairs** or needed additional conversion code that was likely incorrect.

### Problem 2: Distribution Shift in h Values
- **Original training data**: `log2(h)` range = `[1.0, 22.6]`, min h ≈ 2.0
- **Augmented PKL data**: `log2(h)` range = `[3.9, 28.5]`, min h ≈ 15.1

The augmented set contains **zero samples near h = 1** (the mathematical minimum). This biases the model toward predicting large values, inflating error on test samples with small heights.

### Problem 3: Extreme Outliers Distort MSE Loss
The augmented data has a max h of ~375 million (log₂ ≈ 28.5) vs the original max of ~6.45 million (log₂ ≈ 22.6). Since the loss is MSE in log₂ space, these extreme samples dominate the gradient signal and push the model to fit outliers at the expense of the bulk of the distribution.

### Problem 4: Unbalanced (n, k, m) Coverage
The augmented data has 16,000+ samples for `(9,6,2)` and `(9,6,3)` but only ~8,000 for `(9,4,*)` combos — creating a skewed model that performs well on easy cases but poorly on harder ones.

---

## What This Notebook Does

1. **Demonstrates** the root cause of the format mismatch
2. **Implements** the LP-based m-height algorithm from Roth et al. (2025) using the simpler formulation from Fig. 1 of the paper
3. **Generates** synthetic data in the **correct format** (`[n, k, m, P]` lists + separate heights list)
4. **Filters** outliers and ensures balanced coverage across all 9 `(n,k,m)` combinations
5. **Saves** data ready to drop into the existing training pipeline

## Step 2: The LP Algorithm

The m-height computation follows **Fig. 1** from Roth et al. (2025), which is a simplified version of the original LP from Jiang (2024). The key result is:

$$h_m(\mathcal{C}) = \max_{S \subseteq [n\rangle,\, |S|=m}\; \max_{i \in S}\; \max_{u \in \mathbb{R}^k}\; u \cdot g_i$$
$$\text{subject to: } -\mathbf{1} \leq u \cdot (G)_{\bar{S}} \leq \mathbf{1}$$

Where:
- $G = [I_k \mid P]$ is the systematic generator matrix
- $S$ is an $m$-subset of $\{0, 1, \ldots, n-1\}$
- $\bar{S}$ is the complement of $S$ (size $n - m$)
- $(G)_{\bar{S}}$ is $G$ restricted to columns indexed by $\bar{S}$
- $g_i$ is the $i$-th column of $G$

The algorithm requires solving $m \cdot \binom{n}{m}$ LPs, which for our parameter space ranges from 72 (m=2) to 630 (k=4, m=5) LPs per sample.

````python
def compute_m_height_lp(n, k, m, P):
    """
    Compute the m-height of the analog code with generator matrix G = [I_k | P]
    using the LP algorithm from Roth et al. (2025), Figure 1.

    The m-height is:
        h_m(C) = max over all m-subsets S of {0,...,n-1},
                 max over i in S,
                 max_{u in R^k} u*g_i  s.t.  |(G)_Sbar * u| <= 1

    Args:
        n  : total length of codeword (always 9 in this project)
        k  : dimension of the code (4, 5, or 6)
        m  : the height index (2 <= m <= n-k)
        P  : k x (n-k) matrix — the non-identity part of G

    Returns:
        float: the m-height, >= 1.0, or None if all LPs are infeasible
    """
    # Build the full k x n generator matrix G = [I_k | P]
    G = np.hstack([np.eye(k), P])  # shape (k, n)

    best_h = 1.0

    for S in combinations(range(n), m):
        # Complement of S: the indices NOT in S
        Sbar = [j for j in range(n) if j not in S]

        # Submatrix of G for complement columns: shape (k, n-m)
        G_Sbar = G[:, Sbar]

        # Build the constraint matrix once per S (reused for all i in S)
        # A_ub @ u <= b_ub encodes: -1 <= G_Sbar.T @ u <= 1
        A_ub = np.vstack([G_Sbar.T, -G_Sbar.T])  # shape (2*(n-m), k)
        b_ub = np.ones(2 * (n - m))

        for i in S:
            g_i = G[:, i]  # the i-th column of G

            # LP: maximize g_i^T @ u (linprog minimizes, so negate)
            # Bounds: u is free (no box constraints)
            res = linprog(
                c=-g_i,
                A_ub=A_ub,
                b_ub=b_ub,
                bounds=[(None, None)] * k,
                method='highs'
            )

            if res.status == 0:  # optimal solution found
                best_h = max(best_h, -res.fun)
            # res.status == 3 => unbounded => h = inf (skip finite case)

    return best_h


# ── VERIFY AGAINST KNOWN TRAINING DATA ────────────────────────────────────────
print('Verifying LP implementation against known training samples:')
print(f'{"Sample":<8} {"True h":>14} {"LP h":>14} {"Match":>8}')
print('-' * 50)
for i in range(8):
    n_s, k_s, m_s, P_s = orig_data[i]
    h_true = orig_heights[i]
    h_lp   = compute_m_height_lp(n_s, k_s, m_s, P_s)
    match  = abs(h_true - h_lp) / h_true < 1e-4
    print(f'{i:<8} {h_true:>14.4f} {h_lp:>14.4f} {str(match):>8}')
````

````python
# ── BENCHMARK GENERATION SPEED ─────────────────────────────────────────────────
print('LP compute time per sample by (k, m):')
print(f'{"(k,m)":<10} {"#LPs":>8} {"Time(s)":>10} {"Samples/hr":>12}')
print('-' * 44)

combos = [(4,2),(4,3),(4,4),(4,5),(5,2),(5,3),(5,4),(6,2),(6,3)]
for k_b, m_b in combos:
    n_b   = 9
    n_lps = m_b * comb(n_b, m_b)
    P_b   = np.random.uniform(-100, 100, (k_b, n_b - k_b))
    t0    = time.time()
    _     = compute_m_height_lp(n_b, k_b, m_b, P_b)
    dt    = time.time() - t0
    sph   = int(3600 / dt)
    print(f'({k_b},{m_b})      {n_lps:>8} {dt:>10.3f} {sph:>12,}')
````

## Step 3: Synthetic Data Generation Strategy

Based on the analysis above, good synthetic data must:

1. **Format**: Store as `list [n, k, m, P]` with heights in a **separate list** — exactly like the original training files
2. **Balance**: Generate equal numbers of samples per `(n,k,m)` combination across all 9 parameter sets
3. **P-matrix distribution**: Sample entries uniformly from `[-100, 100]` matching the original data
4. **Outlier control**: Reject samples with `log2(h) > 23` (practical max in training data) and skip infinite heights
5. **h-range coverage**: The natural distribution from random P matrices covers the full log2(h) range well — no additional stratification is needed

### Generation approach
- Sample P entries from `Uniform(-100, 100)` using integer or float values
- Compute exact h using LP
- Reject samples where h is infinite or `log2(h) > MAX_LOG2_H` to control outliers
- Collect until each `(n,k,m)` combo has `TARGET_PER_COMBO` samples

````python
# ── GENERATION CONFIGURATION ──────────────────────────────────────────────────

# All (n, k, m) combos per the project spec
PARAM_COMBOS = [
    (9, 4, 2), (9, 4, 3), (9, 4, 4), (9, 4, 5),
    (9, 5, 2), (9, 5, 3), (9, 5, 4),
    (9, 6, 2), (9, 6, 3),
]

# Target samples per (n,k,m) combo — adjust to control total dataset size
# 5000 per combo * 9 combos = 45,000 samples total (~30 min on CPU)
# For a full run targeting 90,000 samples, set to 10,000 per combo
TARGET_PER_COMBO = 5000   # ← increase for larger dataset
MAX_LOG2_H       = 23.0   # Reject samples with extreme outlier heights
P_RANGE          = 100.0  # P values drawn from [-P_RANGE, P_RANGE]
RANDOM_SEED      = 42

np.random.seed(RANDOM_SEED)
print(f'Generating {TARGET_PER_COMBO} samples per combo × {len(PARAM_COMBOS)} combos '
      f'= {TARGET_PER_COMBO * len(PARAM_COMBOS):,} total samples')
print(f'P values from [-{P_RANGE}, {P_RANGE}], outlier cap: log2(h) <= {MAX_LOG2_H}')
````

````python
def generate_samples_for_combo(n, k, m, target, p_range=100.0, max_log2h=23.0,
                                seed=None, verbose_every=500):
    """
    Generate `target` valid samples for a single (n, k, m) parameter set.

    Sampling strategy:
    - Draw P entries from Uniform(-p_range, p_range)
    - Compute exact m-height via LP
    - Accept if h is finite and log2(h) <= max_log2h

    Returns:
        samples : list of [n, k, m, P] — same format as original training data
        heights : list of float          — same format as original height data
    """
    if seed is not None:
        np.random.seed(seed)

    samples, heights = [], []
    attempts = 0
    t0 = time.time()

    while len(samples) < target:
        attempts += 1

        # Sample P matrix: k rows, (n-k) columns, entries in [-p_range, p_range]
        P = np.random.uniform(-p_range, p_range, size=(k, n - k))

        # Compute exact m-height
        h = compute_m_height_lp(n, k, m, P)

        # Acceptance criteria:
        # 1. h must be finite (not inf)
        # 2. log2(h) must be below the outlier cap
        if np.isinf(h) or h <= 0:
            continue
        if np.log2(h) > max_log2h:
            continue

        # Store in the EXACT FORMAT expected by encode_sample in the training notebook
        samples.append([n, k, m, P])
        heights.append(float(h))

        if len(samples) % verbose_every == 0:
            accept_rate = len(samples) / attempts
            elapsed = time.time() - t0
            eta = (elapsed / len(samples)) * (target - len(samples))
            print(f'  ({n},{k},{m}): {len(samples)}/{target}  '
                  f'accept={accept_rate:.1%}  elapsed={elapsed:.0f}s  eta={eta:.0f}s')

    elapsed = time.time() - t0
    accept_rate = len(samples) / attempts
    print(f'  ({n},{k},{m}) DONE: {len(samples)} samples, '
          f'accept rate={accept_rate:.1%}, time={elapsed:.0f}s')

    return samples, heights


print('generate_samples_for_combo() defined.')
print('Run the next cell to start generation (this will take a while).')
````

````python
# ── MAIN GENERATION LOOP ──────────────────────────────────────────────────────
# This cell generates all synthetic data. Progress is printed every 500 samples.
#
# Expected runtime on Colab CPU (T4 not needed — this is pure scipy):
#   TARGET_PER_COMBO=1000  → ~5 minutes
#   TARGET_PER_COMBO=5000  → ~25 minutes
#   TARGET_PER_COMBO=10000 → ~50 minutes

all_samples = []
all_heights = []

total_start = time.time()
print(f'Starting generation at {time.strftime("%H:%M:%S")}')
print('=' * 60)

for combo_idx, (n, k, m) in enumerate(PARAM_COMBOS):
    print(f'\nCombo {combo_idx+1}/{len(PARAM_COMBOS)}: (n={n}, k={k}, m={m})')
    # Use a different seed per combo for independence
    combo_seed = RANDOM_SEED + combo_idx * 1000

    s, h = generate_samples_for_combo(
        n, k, m,
        target=TARGET_PER_COMBO,
        p_range=P_RANGE,
        max_log2h=MAX_LOG2_H,
        seed=combo_seed,
        verbose_every=500
    )
    all_samples.extend(s)
    all_heights.extend(h)

total_elapsed = time.time() - total_start
print(f'\n{"="*60}')
print(f'Generation complete in {total_elapsed/60:.1f} minutes.')
print(f'Total samples: {len(all_samples):,}')
````

## Step 4: Validate the Generated Data

Before saving, verify that the synthetic data:
- Has the correct format expected by `encode_sample`
- Has a well-distributed `log2(h)` range matching the original training data
- Has balanced coverage across all `(n,k,m)` combos

````python
# ── FORMAT VALIDATION ─────────────────────────────────────────────────────────
print('=== FORMAT VALIDATION ===')

# 1. Test that encode_sample works on synthetic data
def encode_sample(sample):
    """Exact copy of encode_sample from the training notebook."""
    n, k, m, P = sample   # ← This is the line that failed with dict format
    p_flat   = P.flatten()
    p_padded = np.zeros(20)
    p_padded[:len(p_flat)] = p_flat
    p_padded = p_padded / 100.0
    features = np.concatenate([[n / 9.0, k / 9.0, m / 9.0], p_padded])
    return features.astype(np.float32)

print('Testing encode_sample on first 10 synthetic samples...')
errors = 0
for i, s in enumerate(all_samples[:10]):
    try:
        feat = encode_sample(s)
        assert feat.shape == (23,), f'Expected shape (23,), got {feat.shape}'
        print(f'  Sample {i}: OK — shape={feat.shape}, n={s[0]}, k={s[1]}, m={s[2]}')
    except Exception as e:
        print(f'  Sample {i}: FAIL — {e}')
        errors += 1

print(f'\nFormat check: {10-errors}/10 passed')

# 2. Check that heights are >= 1.0
h_arr = np.array(all_heights)
print(f'\nAll heights >= 1.0: {(h_arr >= 1.0).all()}')
print(f'Any NaN: {np.isnan(h_arr).any()}')
print(f'Any Inf: {np.isinf(h_arr).any()}')
````

````python
# ── DISTRIBUTION ANALYSIS ─────────────────────────────────────────────────────
h_arr = np.array(all_heights)
log2h = np.log2(h_arr)

print('=== DISTRIBUTION ANALYSIS ===')
print(f'Total samples:   {len(all_samples):,}')
print(f'log2(h) range:   [{log2h.min():.2f}, {log2h.max():.2f}]')
print(f'log2(h) mean:    {log2h.mean():.2f}')
print(f'log2(h) std:     {log2h.std():.2f}')

# Compare to original
orig_log2h = np.log2(np.array(orig_heights))
print(f'\nOriginal log2(h) range:  [{orig_log2h.min():.2f}, {orig_log2h.max():.2f}]')
print(f'Synthetic log2(h) range: [{log2h.min():.2f}, {log2h.max():.2f}]')

# (n,k,m) balance
print(f'\n{"(n,k,m)":<15} {"Synthetic":>10} {"Original":>10}')
synth_counts = Counter((s[0], s[1], s[2]) for s in all_samples)
for key in sorted(synth_counts.keys()):
    orig_cnt = Counter((s[0], s[1], s[2]) for s in orig_data)[key]
    print(f'{str(key):<15} {synth_counts[key]:>10} {orig_cnt:>10}')
````

````python
# ── VISUALIZATION ─────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1. log2(h) histogram comparison
ax = axes[0]
ax.hist(orig_log2h, bins=40, alpha=0.6, color='steelblue', label='Original', density=True)
ax.hist(log2h,      bins=40, alpha=0.6, color='darkorange', label='Synthetic', density=True)
ax.set_xlabel('log₂(h)')
ax.set_ylabel('Density')
ax.set_title('Height Distribution Comparison')
ax.legend()
ax.grid(True, alpha=0.3)

# 2. Per (n,k,m) combo breakdown
ax = axes[1]
combo_labels = [str(c) for c in sorted(synth_counts.keys())]
synth_vals   = [synth_counts[k] for k in sorted(synth_counts.keys())]
orig_vals    = [Counter((s[0],s[1],s[2]) for s in orig_data)[k]
                for k in sorted(synth_counts.keys())]
x = np.arange(len(combo_labels))
ax.bar(x - 0.2, orig_vals,  0.4, label='Original',  color='steelblue', alpha=0.8)
ax.bar(x + 0.2, synth_vals, 0.4, label='Synthetic', color='darkorange', alpha=0.8)
ax.set_xticks(x)
ax.set_xticklabels(combo_labels, rotation=45, ha='right', fontsize=7)
ax.set_ylabel('Sample Count')
ax.set_title('(n,k,m) Coverage Balance')
ax.legend()
ax.grid(True, alpha=0.3)

# 3. log2(h) per combo (synthetic)
ax = axes[2]
log2h_by_combo = defaultdict(list)
for s, h_val in zip(all_samples, all_heights):
    log2h_by_combo[(s[0], s[1], s[2])].append(np.log2(h_val))
means = [np.mean(log2h_by_combo[k]) for k in sorted(log2h_by_combo.keys())]
stds  = [np.std(log2h_by_combo[k])  for k in sorted(log2h_by_combo.keys())]
ax.bar(x, means, yerr=stds, color='darkorange', alpha=0.8, capsize=4)
ax.set_xticks(x)
ax.set_xticklabels(combo_labels, rotation=45, ha='right', fontsize=7)
ax.set_ylabel('log₂(h) mean ± std')
ax.set_title('Synthetic: Height by (n,k,m)')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('synthetic_data_analysis.png', dpi=120, bbox_inches='tight')
plt.show()
print('Figure saved to synthetic_data_analysis.png')
````

## Step 5: Save in Correct Format

Save two separate files — identical in structure to the original training data:
- `synthetic_training_data.pkl` → list of `[n, k, m, P]` (same as `CSCE-636-Project-1-Train-n_k_m_P`)
- `synthetic_training_heights.pkl` → list of floats (same as `CSCE-636-Project-1-Train-mHeights`)

````python
# ── SAVE THE SYNTHETIC DATA ────────────────────────────────────────────────────

# Shuffle before saving to avoid ordering artifacts during training
combined = list(zip(all_samples, all_heights))
np.random.seed(RANDOM_SEED)
np.random.shuffle(combined)
shuffled_samples = [x[0] for x in combined]
shuffled_heights = [x[1] for x in combined]

# Save data file
with open(OUTPUT_SYNTH_DATA_PATH, 'wb') as f:
    pickle.dump(shuffled_samples, f)
print(f'Saved samples to: {OUTPUT_SYNTH_DATA_PATH}')

# Save heights file
with open(OUTPUT_SYNTH_H_PATH, 'wb') as f:
    pickle.dump(shuffled_heights, f)
print(f'Saved heights to: {OUTPUT_SYNTH_H_PATH}')

# ── VERIFY ROUND-TRIP ─────────────────────────────────────────────────────────
print('\nVerifying round-trip load...')
with open(OUTPUT_SYNTH_DATA_PATH, 'rb') as f:
    verify_data = pickle.load(f)
with open(OUTPUT_SYNTH_H_PATH, 'rb') as f:
    verify_heights = pickle.load(f)

print(f'Loaded {len(verify_data)} samples and {len(verify_heights)} heights')
assert len(verify_data) == len(verify_heights), 'Length mismatch!'
# Test encode_sample on loaded data
feat = encode_sample(verify_data[0])
assert feat.shape == (23,)
print(f'encode_sample test on loaded data: OK (shape={feat.shape})')
print('\nAll checks passed. Data is ready for training.')
````

## Step 6: How to Use the Synthetic Data in Your Training Notebook

The synthetic data files drop directly into the existing training pipeline. There are two ways to use them:


### Option B: Merge synthetic with original (recommended)
Add this to your training notebook after loading the original data:

### Why this works (what was wrong before)
The original augmented data stored dicts `{'n':..., 'k':..., 'm':..., 'P':..., 'h':...}`. When the training code did `n, k, m, P = sample`, Python tried to unpack the **keys** of the dict (`'n'`, `'k'`, `'m'`, `'P'`, `'h'`), which is 5 values — causing `ValueError: too many values to unpack`. The synthetic data generated here stores plain lists `[n, k, m, P]` which unpack correctly into exactly 4 variables.

## Summary

### Problems found in `augmented_training_data_300k.pkl`

| Problem | Description | Impact |
|---------|-------------|--------|
| **Format mismatch** | Dict with 5 keys instead of 4-element list | `ValueError` when `encode_sample` tries `n,k,m,P = sample` |
| **Distribution shift** | min h = 15.1 (no near-1 samples) | Biases model toward large predictions |
| **Extreme outliers** | max log₂(h) ≈ 28.5 vs original 22.6 | Outliers dominate MSE loss, distort gradients |
| **Imbalanced coverage** | (9,6,2) has 2× more samples than (9,4,*) | Model underfits underrepresented combos |

### What the synthetic data fixes

- ✅ **Correct format**: `[n, k, m, P]` list + separate heights list — drops directly into existing pipeline
- ✅ **Balanced coverage**: equal samples per (n,k,m) combination
- ✅ **Matched distribution**: P values from `Uniform(-100, 100)`, log₂(h) range matches original
- ✅ **Outlier control**: hard cap at `log₂(h) ≤ 23` prevents extreme values from dominating loss
- ✅ **Exact LP labels**: heights computed from the same algorithm that generated the original training set
